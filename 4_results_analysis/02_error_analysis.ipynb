# Bin differences in predictions

For now, we can just focus on the differences in predictions between hour 0 and 24. We'll bin the CSNs according to the absolute difference in predictions (`abs_diff_0pred_24pred`).

```python
# create column to mark these 
bins = [x*0.1 for x in range(0,11)]
results['abs_diff_bins'] = pd.cut(results['abs_diff_0pred_24pred'], bins)
view_df(results)
```

Binned the absolute difference between prediction at 0 and 24 hours. checking how these differences stack up when broken up by the differece in True labels. 

For the true labels, the difference between 0 and 24 hours is either:

- -1 = 0-1 = non-ICU to ICU
- 0 = 0-0 or 1-1 = non-ICU to non-ICU, or ICU to ICU
- 1 = 1-0 = ICU to non-ICU

```python
# difference in predictions (binned) vs. difference in True labels (either -1,0,1)
diffs = results.groupby(['abs_diff_bins', 'diff_0True_24True']).size().unstack(fill_value=0)
diffs['sum'] = diffs.sum(axis=1)
diffs
```

```python
# looking at percentages for the table above
perc = diffs.loc[:,-1:1].div(diffs['sum'], axis=0) * 100
perc = perc.round(2)
perc
```

```python
# mark the events with some labels that make things easier later
adt_admit = adt_admit.sort_values(['pat_enc_csn_id_coded', 'seq_num_in_enc'])

# **assumes dataframe is sorted by time

# mark whether previous entry had pat class Emergency Services
adt_admit['prev_emerg'] = adt_admit.pat_class.shift() == 'Emergency Services'
# mark whether current event has pat class Inpatient
adt_admit['curr_inpatient'] = adt_admit.pat_class == 'Inpatient'
# mark whether current event is continued csn as previous
adt_admit['continued'] = adt_admit.pat_enc_csn_id_coded.eq(
    adt_admit.pat_enc_csn_id_coded.shift())

# find our cases by taking the AND of these columns
cols = ['prev_emerg', 'curr_inpatient', 'continued']
adt_admit['first_ip'] = adt_admit[cols].all(axis=1)

adt_admit.head()
```

# Pat lv of care

Look at pat lv of care when patients are admitted

```python
adt_admit.columns
```

```python
# change admit time and effective time to datetime - these take a littl while to run!!

# change the effective time to datetime since read in from csv
adt_admit.effective_time_jittered_utc = pd.to_datetime(adt_admit.effective_time_jittered_utc)

# change admit time
adt_admit.admit_time = pd.to_datetime(adt_admit.admit_time, utc=True)
```

```python
adt_admit.head()
```

```python
# filter down to 24hrs after admit
adt_admit['time_since_admit'] = adt_admit.effective_time_jittered_utc - adt_admit.admit_time

adt_results_24hr = adt_admit[adt_admit.time_since_admit < timedelta(hours=24)]

adt_results_24hr.head()
```

```python
# filter to inpatient events after admit time
ip_events = adt_results_24hr[(adt_results_24hr.admit_time <= adt_results_24hr.effective_time_jittered_utc) &
                     (adt_results_24hr.pat_class == 'Inpatient')]

ip_events.pat_enc_csn_id_coded.nunique()
```

```python
# keep only change of lv of care status

# sort by sequence 
ip_events = ip_events.sort_values(['pat_enc_csn_id_coded', 'seq_num_in_enc'])

# keep only those that change
change_care = ip_events[~(ip_events.pat_lv_of_care.eq(ip_events.pat_lv_of_care.shift()) &
                         ip_events.pat_enc_csn_id_coded.eq(ip_events.pat_enc_csn_id_coded.shift()))]

change_care.pat_enc_csn_id_coded.nunique()
```

```python
change_care.size
```

```python
change_care[change_care.pat_enc_csn_id_coded.eq(change_care.shift().pat_enc_csn_id_coded)].head()
```

```python
trajectory = change_care[['pat_enc_csn_id_coded', 'pat_lv_of_care']].groupby('pat_enc_csn_id_coded').agg(
{'pat_lv_of_care' : lambda x: ' -> '.join([str(c) for c in x])}).reset_index().rename({'pat_lv_of_care': 'care_trajectory'}, axis=1)
```

```python
# join to the change care df again
ip_traj = trajectory.merge(ip_events[['pat_enc_csn_id_coded', 'abs_diff_bins']], how='left').drop_duplicates()
```

```python
ip_traj.columns
```

```python
counts = ip_traj.groupby(['abs_diff_bins', 'pat_enc_csn_id_coded']).care_trajectory.value_counts().groupby(['abs_diff_bins', 'care_trajectory']).describe()
```

```python
traj_count = ip_traj.groupby(['abs_diff_bins']).agg(traj_count=pd.NamedAgg(column='care_trajectory', aggfunc='value_counts')).reset_index().sort_values(['abs_diff_bins', 'traj_count'], ascending=False)

traj_count
```

```python
traj_count_sub = traj_count.groupby('abs_diff_bins').head(10)

trajectory_file = savedir + "02_cohort_care_trajectories.tsv"
traj_count_sub.to_csv(trajectory_file,
                     sep="\t", index=False)
```
